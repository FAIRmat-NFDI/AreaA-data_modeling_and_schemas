```python
import xml.etree.ElementTree as ET

def parse_raman_spectrometer_file(file_path):
    tree = ET.parse(file_path)
    root = tree.getroot()

    metadata = {}
    measured_data = {}

    # Extract metadata
    def extract_metadata(root):
        metadata_dict = {}
        for elem in root.iter('LSX'):
            if elem.text:
                parent = elem.getparent()
                if parent is not None and parent.tag == 'LSX' and 'Format' in parent.attrib:
                    key = elem.text
                    value = parent.find("LSX[@Format='5']").text if parent.find("LSX[@Format='5']") is not None else None
                    metadata_dict[key] = value
        return metadata_dict

    # Traverse and extract relevant metadata
    lsx_tree = root.find('.//LSX_Tree')
    metadata['Title'] = lsx_tree.find(".//LSX[@Format='8']").text if lsx_tree.find(".//LSX[@Format='8']") is not None else None
    date_element = next((e for e in lsx_tree.iter('LSX') if e.text == 'Date'), None)
    metadata['Date'] = date_element.find("../LSX[@Format='7']").text if date_element is not None else None
    acquired_element = next((e for e in lsx_tree.iter('LSX') if e.text == 'Acquired'), None)
    metadata['Acquisition Date'] = acquired_element.find("../LSX[@Format='7']").text if acquired_element is not None else None

    # Extract other metadata
    metadata.update(extract_metadata(lsx_tree))

    # Extract measured data
    wavenumbers_text = lsx_tree.find('.//LSX[@ID="0x7D6CD4DB"]').text.strip()
    intensities_text = root.find('.//LSX_Matrix').find('.//LSX_Row').text.strip()

    measured_data['wavenumbers'] = list(map(float, wavenumbers_text.split()))
    measured_data['intensities'] = list(map(float, intensities_text.split()))

    return metadata, measured_data

file_path = '3611subs.xml'
metadata, measured_data = parse_raman_spectrometer_file(file_path)

# Print the metadata and first few measured data points
print("Metadata:")
for key, value in metadata.items():
    print(f"{key}: {value}")

print("\nFirst 10 measured data points:")
print("Wavenumbers:", measured_data['wavenumbers'][:10])
print("Intensities:", measured_data['intensities'][:10])
```

```python
import xml.etree.ElementTree as ET

# Function to print the data in the specified format
def print_data(element, level=0):
    indent = '\n' + '\t' * level
    print(f"{indent}Format: {element.get('Format', 'None')},")
    print(f"{indent}ID: {element.get('ID', 'None')},")
    print(f"{indent}Size: {element.get('Size', 'None')},")
    value = element.text.strip() if element.text else ''
    print(f"{indent}Value: '{value}',")
    children = list(element)
    print(f"{indent}Children: [", end='')
    if children:
        for child in children:
            print_data(child, level + 1)
            print(f"{indent}\t", end='')
    print(f"{indent}]", end='')

# Read and parse the XML file
tree = ET.parse('3611subs.xml')
root = tree.getroot()

# Print the structured data starting from the root element
print_data(root)
```
